```
import datetime as dt
import pandas as pd
import pandas_datareader.data as web
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.style as style
%matplotlib inline
```

```
start = dt.date(2000,1,1)
end = dt.date(2020,5,10)
source = 'yahoo'
company = 'YDUQ3.SA'
data = web.DataReader(company,source,start,end)
```

```
data.tail(5)
```

```
SMA30 = pd.DataFrame()
SMA30['Adj Close'] = data['Adj Close'].rolling(window=30).mean()
```

```
SMA100 = pd.DataFrame()
SMA100['Adj Close'] = data['Adj Close'].rolling(window=100).mean()
```

```
style.available
```

```
style.use('seaborn-paper')
plt.figure(figsize=(12.5,8))
plt.plot(data['Adj Close'])
plt.plot(SMA30['Adj Close'])
plt.plot(SMA100['Adj Close'])
plt.legend(['Adj Close','SMA30','SMA100'])
plt.show()
```
